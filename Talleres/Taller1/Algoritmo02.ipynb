### **Algoritmo02**

```python
def alg_02(a):
    n = len(a)
    count = 0
    for i in range(n):
        swapped = False
        for j in range(1, n - i):
            if a[j] < a[j - 1]:  # Comparaci√≥n de elementos adyacentes
                a[j], a[j - 1] = a[j - 1], a[j]  # Intercambio de valores
                swapped = True
                count += 1  # Contador de intercambios
        if not swapped:  # Si no hubo intercambios, la lista est√° ordenada
            break
    return a, count  # Retorna la lista ordenada y el n√∫mero de intercambios
```

##### **Respuesta ->** Ordena el vector

```python
# Caso de prueba
lista = [5, 4, 3, 2, 1]
print("Lista original:", lista)
print("Lista ordenada y el n√∫mero de comparaciones:", alg_02(lista))
```
