##### Algoritmo01

```python
def alg_01(N, X):
    SUM = 0
    for i in range(N):
        SUM += X[i]
    return SUM
```

```python
# Casos de prueba
X1 = list(range(100))  # Caso 1
print("Resultado caso 1:", alg_01(len(X1), X1))

X2 = [1, 1/2, 1/4]  # Caso 2
print("Resultado caso 2:", alg_01(len(X2), X2))

X3 = [1/2**x for x in range(15)]
print(X3)
r = alg_01(len(X3), X3)
print(r)
f"{2-r}"
```

##### **Respuesta ->** El limite de la serie es 2

```python
# Resolucion quizz
X2 = [1/2**x for x in range(6)]  # 5 elementos como minimo para que el error relativo este dentro de 10^-1
X4 = round(alg_01(len(X2), X2))
error_relativo = abs( alg_01(len(X2), X2) - X4)/ abs(alg_01(len(X2), X2))
print(error_relativo)
```
