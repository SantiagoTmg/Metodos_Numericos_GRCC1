##### Algoritmo01

```python
def alg_01(N, X):
    SUM = 0
    for i in range(N):
        SUM += X[i]
    return SUM
```

```python
# Casos de prueba
X1 = list(range(100))  # Caso 1
print("Resultado caso 1:", alg_01(len(X1), X1))

X2 = [1, 1/2, 1/4]  # Caso 2
print("Resultado caso 2:", alg_01(len(X2), X2))

X3 = [1/2**x for x in range(15)]
print(X3)
r = alg_01(len(X3), X3)
print(r)
f"{2-r}"
```
