#### **Algoritmo 04**

```python
def serie_geometrica(n):
    suma = 0
    for i in range(1, n + 1):
        suma += 1/i
    return suma
```

### **¿A que valor converge el algoritmo 04?**

```python
for i in range(1, 25):
    print(f"{i}: {serie_geometrica(i)}")
```

#### **Respuesta ->** El algoritmo tiende al infinito
