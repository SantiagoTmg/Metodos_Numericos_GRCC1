#### **Algoritmo 03**

```python
import math

def alg_03(n):
    if n == 0:
        return 0
    else:
        x = 0
        y = 1
        for i in range(1, n):
            z = x + y
            x = y
            y = z
        return y
```

```python
def alg_03_optimizado(n):
    if n < 2:
        return n
    x, y = 0, 1
    for _ in range(n - 1):
        x, y = y, x + y  # Optimizaci√≥n en la actualizaci√≥n
    return y
```

```python
def FL_rounding(x):
    if x == 0:
        return 0
    else:
        return round(x, 3)  # Redondeo a 3 decimales
```

```python
import time
start_t = time.time()
[alg_03(x) for x in range(5_000)]
end_t = time.time()
end_t - start_t
```

```python
start_t = time.time()
[alg_03_optimizado(x) for x in range(5_000)]
end_t = time.time()
end_t - start_t
```

```python
# Soluci√≥n quiz 7
constante = (1 + math.sqrt(5)) / 2
y_0 = 1 # y anterior
y_1 = alg_03(1) # y actual
iteracion = 1 # Iteraci√≥n inicial

while iteracion < 1000:
    # Redondear a 3 decimales
    y_0_redondeado = FL_rounding(y_0)
    y_1_redondeado = FL_rounding(y_1)

    cociente = y_1_redondeado / y_0_redondeado
    # Se redondea el cociente a 3 decimales
    cociente_redondeado = FL_rounding(cociente)

    error_relativo = abs(constante - cociente_redondeado) / constante

    if error_relativo < 1e-5:
        break

    iteracion += 1
    y_0 = y_1
    y_1 = alg_03(iteracion)

# Como el n√∫mero de iteraciones es igual a 1000, se puede decir que no hay iteracion en la que el error relativo este dentro de 10^-5
print(f"El n√∫mero de iteraciones necesarias para que el error relativo este dentro de 10^-5 es: {iteracion}")
```
