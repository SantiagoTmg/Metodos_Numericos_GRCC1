#### **Algoritmo 03**

```python
def alg_03(n):
    if n == 0:
        return 0
    else:
        x = 0
        y = 1
        for i in range(1, n):
            z = x + y
            x = y
            y = z
        return y
```

```python
# Caso de prueba
n = 11
print("Y es igual a:", alg_03(n))
```
