#### **Ecuación de tercer grado**

image.png

```python
import numpy as np
import matplotlib.pyplot as plt

# Define la ecuación
def ecuacion(x: float) -> float:
  s = 1.0/4*(x**3 + 3*x**2 - 6*x -8)
  return s

# Generacion de valores de x
x = np.linspace(-10, 10, 100)

# Calcula los valores de y
y = ecuacion(x)

# Plot the equation
plt.plot(x, y)

plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot de $$')

ax = plt.gca()
ax.set_ylim([-10, 10])
ax.set_xlim([-5, 3])
plt.grid(True)
plt.show()
```

#### **Bisección**

```python
def biseccion(ecuacion, a, b, TOL, N0):
    """
    Implementa el método de bisección para encontrar la raíz de f(x).

    Parámetros:
    ecuacion    - Función a evaluar.
    a, b - Extremos iniciales del intervalo.
    TOL  - Tolerancia para la precisión.
    N0   - Número máximo de iteraciones.
    """
    FA = ecuacion(a)
    for i in range(1, N0 + 1):
        p = a + (b - a) / 2
        FP = ecuacion(p)

        if FP == 0 or (b - a) / 2 < TOL:
            print(f"Raíz encontrada: p = {p}, iteración {i}")
            return p

        if FA * FP > 0:
            a = p
            FA = FP
        else:
            b = p

    print(f"El método falló después de {N0} iteraciones.")
    return None


a = -5  # Extremo inferior
b = 3  # Extremo superior
TOL = 1e-5  # Tolerancia
N0 = 50     # Número máximo de iteraciones

raiz = biseccion(ecuacion, a, b, TOL, N0)
```
