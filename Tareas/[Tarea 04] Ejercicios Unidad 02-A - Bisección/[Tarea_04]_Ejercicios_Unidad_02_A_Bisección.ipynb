#### **1. Use el método de la bisección para encontrar soluciones precisas dentro de 10^(-2) para $$x^3 - 7x^2 + 14x - 6 = 0$$ en cada intervalo.**

```python
import numpy as np

# Definimos la ecuación
def ecuacion(x):
    return x**3 - 7*x**2 + 14*x - 6

# Implementación optimizada del método de bisección
def biseccion(ecuacion, a, b, TOL, N0):
    """
    Método de bisección con condición mejorada para reducir iteraciones.

    Parámetros:
    ecuacion - Función a evaluar.
    a, b - Extremos iniciales del intervalo.
    TOL  - Tolerancia para la precisión.
    N0   - Número máximo de iteraciones.
    """
    FA = ecuacion(a)
    p = a  # Inicializamos p fuera del bucle para compararlo después

    for i in range(1, N0 + 1):
        p_anterior = p  # Guardamos el punto anterior
        p = (a + b) / 2
        FP = ecuacion(p)

        if FP == 0 or abs(p - p_anterior) < TOL:  # Detener si el cambio es menor que la tolerancia
            print(f"Raíz encontrada: p = {p:.2f}, iteración {i}")
            return p

        if FA * FP > 0:
            a = p
            FA = FP
        else:
            b = p

    print(f"El método falló después de {N0} iteraciones.")
    return None
    
```

##### a.- [0, 1]

```python
import matplotlib.pyplot as plt
# Definición del intervalo, tolerancia y número de iteraciones
a, b = 0, 1
TOL = 10**-2
N0 = 50

# Ejecutar el método de bisección
root = biseccion(ecuacion, a, b, TOL, N0)

# Generar la gráfica de la función en el intervalo [0, 1]
x_values = np.linspace(a, b, 500)
y_values = ecuacion(x_values)
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label="f(x) = x^3 - 7x^2 + 14x - 6", color='blue')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(root, color='red', linestyle="--", label=f"Root ≈ {root:.2f}")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Gráfica de la función y aproximación de la raíz por el método de bisección")
plt.legend()
plt.grid(True)
plt.show()
```

#### **1. 	Un abrevadero de longitud L tiene una sección transversal en forma de semicírculo con radio r. (Consulte la figura adjunta.) Cuando se llena con agua hasta una distancia h a partir de la parte superior, el volumen V de agua es $$V = L \left[ 0.5 \pi r^2 - r^2 \arcsin \left( \frac{h}{r} \right) - h \sqrt{r^2 - h^2} \right]$$**

#### **Suponga que L=10 cm,r=1cm y V=12.4cm^3. Encuentre la profundidad del agua en el abrevadero dentro de 0.01 cm.**

```python
def ecuacion_h(h):
    r = 1  # Radio del abrevadero
    L = 10 # Longitud del abrevadero
    V = 12.4 # Volumen dado
    return L * (0.5 * np.pi * r**2 - r**2 * np.arcsin(h/r) - h * np.sqrt(r**2 - h**2)) - V

a, b = 0, 1  # La profundidad está entre 0 y el radio
TOL = 0.01
N0 = 50

h_aproximada = biseccion(ecuacion_h, a, b, TOL, N0)

if h_aproximada is not None:
    print(f"La profundidad aproximada del agua es h ≈ {h_aproximada:.2f} cm")
```

#### **2. 	Un objeto que se cae verticalmente a través del aire está sujeto a una resistencia viscosa, así como a la fuerza de gravedad. Suponga que un objeto con masa m cae desde una altura s_o y que la altura del objeto después de t segundos es $$s(t) = s_o - \frac{mg}{k} t + \frac{m^2 g}{k^2} \left( 1 - e^{-\frac{kt}{m}} \right)$$**

#### **Donde g=9.81 m/s^2 y k representa el coeficiente de la resistencia del aire en Ns/m. Suponga s_o=300 m, m=0.25 kg y k=0.1 Ns/m. Encuentre, dentro de 0.01 segundos, el tiempo que tarda un cuarto de kg en golpear el piso.**

```python
def ecuacion_s(t):
    s0 = 300   # Altura inicial
    m = 0.25   # Masa en kg
    g = 9.81   # Gravedad en m/s^2
    k = 0.1    # Coeficiente de resistencia del aire
    return s0 - (m*g/k)*t + (m**2*g/k**2)*(1 - np.exp(-k*t/m))

a, b = 0, 30  # Aproximación del tiempo de caída
TOL = 0.01
N0 = 50

tiempo_aproximado = biseccion(ecuacion_s, a, b, TOL, N0)

if tiempo_aproximado is not None:
    print(f"El tiempo aproximado de caída es t ≈ {tiempo_aproximado:.2f} segundos")
```

#### **1. 	Use el teorema 2.1 para encontrar una cota para el numero de iteraciones necesarias para lograr una aproximación con precisión de 10^(-4) para la solución de x^3-x-1=0 que se encuentra dentro del intervalo [1, 2]. Encuentre una aproximación para la raíz con este grado de precisión.**

```python
def ecuacion_c(x):
    return x**3 - x - 1

# Definimos el intervalo y los parámetros
a, b = 1, 2
TOL = 10**-4
N0 = 14

# Lista para almacenar los puntos medios
puntos_medios = []

# Ejecutamos el método de bisección y almacenamos los puntos medios
def biseccionTeorico(ecuacion, a, b, TOL, N0):
    FA = ecuacion(a)
    p = a

    for i in range(1, N0 + 1):
        p_anterior = p
        p = (a + b) / 2
        FP = ecuacion(p)
        puntos_medios.append(p)  # Guardamos el punto medio externo a la función

        if FP == 0 or abs(p - p_anterior) < TOL:
            return p

        if FA * FP > 0:
            a = p
            FA = FP
        else:
            b = p

    return None

# Ejecutamos el método de bisección
raiz_aproximada = biseccionTeorico(ecuacion_c, a, b, TOL, N0)

# Generamos la gráfica
x_values = np.linspace(0, 2.5, 500)
y_values = ecuacion_c(x_values)

plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label="f(x) = x³ - x - 1", color="blue")
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(raiz_aproximada, color="green", linestyle="--", label=f"Raíz aproximada: x ≈ {raiz_aproximada:.5f}")

# Marcar los puntos medios en la gráfica
plt.scatter(puntos_medios, ecuacion(np.array(puntos_medios)), color="red", label="Puntos medios", zorder=3)

plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Método de Bisección aplicado a f(x) = x³ - x - 1")
plt.legend()
plt.grid(True)
plt.show()
```
